## Decision Tree Classification Algorithm

- **Decision Tree** is a Supervised learning technique that can be used for both:
    - Classification
    - Regression problems
    - But mostly it is preferred for solving **Classification** problems. 

 - **Decision tree** is called a decision tree because, similar to a tree, it starts with the root node, which expands on further branches and constructs a tree-like structure.
 
### Building Decision Tree

 - Building a tree on a given labeled dataset can be achieved using the **CART algorithm**, which stands for **Classification and Regression Tree algorithm**.
 
 
 - The concept of **CART** is straightforward, because a classification tree simply learns by answering a sequence of **if-else or (Yes/No)** questions about individual features to infer the labels, it also splits the tree into subtrees based on that answer.
 
 
### Why use Decision Trees?

 - **Non Linearity**: In contrast to linear models, decision tree has the ability to capture non-linear relationships between features and labels.
 
 - **Easiness**: decision Trees usually mimic human thinking ability while making a decision, so it is easy to understand.
 
 - **Trees** don't require the features to be on the same scale through standardization for example; however, data preprocessing may improve the accuracy of trees.
 


## Decision Tree Terminologies

**Decision Tree** is a tree-structured classifier, consisting of a hierarchy of individual units called nodes.

- **Node** is a point that involves either a question or a prediction. 

- **Root Node**: is from where the decision tree starts. It represents the entire dataset, which further gets divided into two or more homogeneous sets.
 
- **Internal nodes** represent the features of a dataset, and they are the decision nodes that are used to make any decision and have multiple branches.

- **Branches** represent the decision rules.

- **Splitting**: is the process of dividing the decision node/root node into sub-nodes according to the given conditions.


- **Sub-Tree**: A tree formed by splitting the tree.

- **Parent/Child node**: The root node of the tree is called the **parent node**, and other nodes are called the **child nodes**.


- **Leaf Node (Terminal node)**: 
    - Leaf nodes are the final output node.
    - A leaf node has one parent but no children.
    - At leaf node, the tree cannot be segregated further after getting a leaf node.
    - **Leaf nodes** are the output of the decision nodes and do not contain any further branches.
    - **Leaf nodes** is where the prediction is made.
    

- **Pruning**: it is the process of removing the unwanted branches from the tree.

- **Max-depth**: the distance between the root node (depth = 0) and the final leaf.

## Classification Decision Tree Workflow

When a classification tree is trained on a dataset, the tree learns a sequence of **if-else (yes-no)** questions with each question involving one feature and one split-point. This process continues until the tree reaches the end where the predicted label will take place. The next steps will walk through the steps of building a tree:


 - **Step 1**: Begin the tree with the root node, says S, which contains the complete dataset.
 
 
 - **Step 2**: Find the best attribute in the dataset using Attribute Selection Measure (ASM).
 
 
 - **Step 3**: Divide the S into subsets that contains possible values for the best attributes.
 
 
 - **Step 4**: Generate the decision tree node, which contains the best attribute.
 
 
 - **Step 5**: Recursively make new decision trees using the subsets of the dataset created in step 3. 
 
Continue this process until a stage is reached where you cannot further classify the nodes and called the final node as a leaf node.

### Decision Regions

- A classification-model divides the feature-space into regions where all instances in one region are assigned to only one class-label. These regions are known as decision-regions.

- **Classification trees** produce rectangular decision-regions in the feature-space. This happens because at each split made by the tree, only one feature is involved. 

- While **Linear models** such logistic classifier produces a line (the boundary is a **straight-line**) that separates the labels.

## Advantages and Disadvantages of CARTs

### Advantages of CARTs: 

 - **Classification and Regression Trees (CARTs)** present many advantages, we mention few:
     - They are easy to understand and their output is easy to interpret.
     - CARTs are easy to use and their flexibility gives them an ability to describe nonlinear dependencies between features and labels. 
     - CARTs don't need a lot of feature preprocessing such as standardization and normalization.
     
### Limitations of CARTs

- A classification tree for example, is only able to produce orthogonal decision boundaries. 

- CARTs  suffer from **high variance (known as overfitting)** when they are trained without constraints; in other words, they are very sensitive to small variations in the training set, Sometimes, when a single point is removed from the training set, a CART's learned parameters may changed drastically. 

## Attribute Selection Measures

 - When a classification tree is trained on a labeled dataset, the tree learns patterns from the features in such a way to produce the purest leaf nodes. In other words the tree is trained in such a way so that, in each leaf, one class-label is predominant. But, **how can this happen?**

 - In linear models, we have seen different statistical measures to assess the model. So,  **What are the criteria is used to measure the impurity of a node?**. In fact, there are different criteria that we can use such as **gini-index**, and **entropy** for **information gain**. 
 

### 1. Entropy

 - **Entropy** is an information theory metric. In data science, the entropy is used as a way to measure how **mixed** a column is, or the **impurity of an attribute**. for a deeper understanding of this concept, check [here](https://en.wikipedia.org/wiki/Information_theory#Quantities_of_information)
 
 
### $$ Entropy (E)= - \sum_{i=1}^{k}P(X_i)\log_2 P(X_i)$$

**Where**,

 - **K**: number of classes
 - **log base 2**: it is important to use $\log_2$. 
 - **P(Xi)**: is the percent of the group that belongs to a given class.


### Or (Litterally)

$$\textbf{Entropy}(S)= -P(yes)\log_2 P(yes)- P(no) \log_2 P(no)$$


**Where**,

  - **S** = Total number of samples. 
  - **P(yes)** = probability of yes
  - **P(no)** = probability of no.



### 2. Information Gain


$$\textbf{IG}\big ( \underbrace{T}_\text{feature}, \underbrace{sp}_\text{split point} \big) = \textbf{I}\textbf{(parent)} - \Bigg (\frac{N_{left}}{N} I(left) + \frac{N_{right}}{N} I(right) \Bigg) $$

**Where**, 

  - $\textbf{I}\textbf{(parent)}$: is the impurity measure of the parent node.
  - $\mathcal{N}$: the number of data points.
  - **sp**: the attribute to split on.
  - **I(left)**: impurity of the left node.
  - **N(left)**: number of data points in the left node.
  - **I(right)**: impurity of the right node.
  - **N(right)**: number of data points in the right node.
  
  
**Here is the formula of information gain with entropy index**

### $$\textbf{IG}(T, A) = \textbf{Entropy(T)} - \sum_{v \in A}\frac{|T_{v}|}{T} \textbf{Entropy}(T_{v}) $$

**which can be understood as**,

#### $$\textbf{Information Gain}= \textbf{Entropy}(T)- \big [\textbf{(Weighted Avg)} * \textbf{Entropy(each feature)}\big]  $$



#### Where,

 - **T**: Target variable
 - **A**: Attribute to split on (the variable to test splitting on).
 - **v**: value from attribute **A** 
 - **T_v**: a subset of the total points in **T**, only unique values in this subset.

### 3. Gini Index


- Gini index is a measure of impurity or purity used while creating a decision tree in the CART algorithm.

- An attribute with the **low Gini index** should be preferred as compared to the **high Gini index**.


- It only creates binary splits, and the CART algorithm uses the Gini index to create binary splits.


Gini index can be calculated using the formula below:


### $$\textbf{Gini Index (GI)} = 1- \sum_{j}^{k}\big (p_{j}\big)^{2}$$

## The Splitting Process of Decision Trees

- The question that you may have in mind is: **Based on the metrics above, how a tree gets split?** 


- The answer to this question needs a deep understanding of **Information Gain**. In this section we will mention the steps of **classification tree splitting process**. 

### Classification-Tree Learning Process

- The CARTs algorithm depends on the **recursion** concept; in other words, the nodes of a classification tree are grown **recursively**. the obtention of an internal node or a leaf depends on the state of its predecessors (the previous nodes).

- The algorithm aims to produce the **purest** (one label class) leaves possible. But **how does the split happen? and which feature (attribute) to split on first**.

#### Maximizing Information Gain

 - The answer to the previous questions exists in **maximizing the information gain** (see the formula above) or **minimizing the gini index**. Here how it works:
 
 
 - The tree considers that every node contains information and aims at **maximizing the Information Gain** obtained after each split. 
 
 
 - The tree tries every feature in the dataset, and the feature that **maximizes** the **information gain** will be picked first, and this process will be repeated at each internal node. (see the workflow above)


- If the **information gain (IG)** obtained by splitting a node is **null**, the node is declared a leaf.

- Keep in mind that these rules are for **unconstrained trees**, which means in practice we can apply constraints on the tree. For example, if you constrain the **maximum depth** of a tree to 3, all nodes having a depth of 3 will be declared leaf nodes even if the **information gain (IG)** obtained by splitting such nodes is not null.


### Pruning: Getting an Optimal Decision tree

- **Pruning** is a process of deleting the unnecessary nodes from a tree in order to get the optimal decision tree.

- A too-large tree increases the risk of **overfitting**, and a small tree may not capture all the important features of the dataset **(underfitting)**. Therefore, a technique that decreases the size of the learning tree without reducing accuracy is known as **Pruning**. There are mainly two types of tree pruning technology used:

  - **Cost Complexity Pruning**.
  - **Reduced Error Pruning**.

  - Here are the most used parameters when training a decision tree classifier.
      - **criterion**: the metric used to measure the quality so to split the features on, **gini** (the default) or **entropy**.

      - **min_samples_split** : The minimum number of samples required to split an internal node. 
      - **min_samples_leaf** : The minimum number of samples required to be at a leaf node. 
      - **max_features**: The number of features to consider when looking for the best split. 
      - **random_state**: setting this parameter will ensure reproducible results.
      - **ccp_alpha**: Complexity parameter used for Minimal Cost-Complexity Pruning.

````python
# Generate predictions
# --------------------

preds = dt.predict(X_test)
````

````python
# Print the confusion matrix and the classification report
# --------------------------
print("*"*54)
acc = accuracy_score(preds, y_test)
print("Test set accuracy: {:.2f}".format(acc))

print("*"*54)
print(confusion_matrix(y_test,preds), "\n")

print("*"*54)
print(classification_report(y_test,preds))
print("*"*54)
````

### Hyperparameter Tuning

- **Recall** that:

    - **Hyperparameter tuning** consists of searching for the set of **optimal hyperparameters** for the learning algorithm. 
    - The solution involves finding the set of **optimal hyperparameters** yielding an optimal model.
    - The optimal model yields an **optimal score**. The score function measures the agreement between true labels and a model's predictions. 

### Note:

 - In sklearn, it defaults to accuracy for classifiers and r-squared for regressors.
 - A model's generalization performance is evaluated using **cross-validation**.
 

### Hyperparameter Tuning  Approaches: 


  - There are many approaches for hyperparameter tuning including:
      - **Grid search**
      - **Random search**
      - **Bayesian optimization**
      - **Gradient-based optimization**
      - **Genetic algorithms**


> In order to get the most out of this course, we will be using different different **hyperparameter tuning** approach for each sub-chapter of this series of courses. We will start with **gridsearch** to tune  decision tree hyperparameter.



## Tuning Decision Tree Using Grid Search Cross Validation

- Here are the steps to tune a decision tree using grid-search cross-validation:
    - Manually set a grid of discrete hyperparameter values.
    - Choose a metric for scoring model performance.
    - Search exhaustively through the grid.
    - Evaluate each model's score for each set of hyperparameters.
    - The optimal hyperparameters are those for which the model achieves the best cross-validation score. 
    
### Note:

  - **Grid Search CV** is computationaly expensive, specification when there are a larger number of features (**curse of dimensionality**). And the bigger the grid the longer it takes to find the optimal solution.

### Performing Grid Search CV in Practice:

 - To understand the steps of this procedure practically. We will use few hyperparameters to tune a decision tree model; for example:
 
     - **The maximum-depth**: `max_depth`
     - **The minimum-percentage of samples per leaf** `min_samples_leaf`. 
     - **The complexity parameter** `ccp_alpha`.
     - **The maximum number of features to consider when splitting**`max_features`
 
- **A K-fold cross-validation** technique is used to evaluate the model performance on each combination of hyperparameters. The optimal hyperparameters correspond to the model achieving the best cross-validation score.

#### The Steps of Hyperparameter Tuning

 - **First**,instantiate a classifier object `DecisionTreeClassifier`.
  - **Create** a dictionary of the hyperpameters you aim to tune, for example
 
         grid_params = {
             'max_depth': [3,4,5,6, 7, 8 , 9],
             'min_samples_leaf': [0.02,0.04, 0.06, 0.08],
             'max_features': [0.2, 0.4,0.6, 0.8, 0.1]
              }
 - **Instantiate** and grid search CV `GridSearchCV`:
     - Set the `estimater` option to your classifier
     - Pass the hyperparameter dict to `param_grid` parameter.
     - Set `scoring` to **accuracy** in case of classification.
     - Set `CV` to a number that suits your needs (5, or 10 are very common).
     - Set the number of cpu's on your machine to `n_jobs` parameter.


###  Extracting the Best Hyperparameters

   - After training the grid search CV model, you can extract results as follows: 
       - **the best set of hyperparameter values** are stored in the attribute `.best_params_`.  
       - **The best cross validation accuracy** can be accessed through `.best_score_` attribute.
       - **The best-model** is stored in `.best_estimator` attribute. (Note that this model is fitted on the whole training set because the `refit` parameter of `GridSearchCV` is set to True by default)
       - 
### Honest Assesment

 - It is best to evaluate the model on unseen data (test set). which is the final step in your workflow of building a machine learning model. 


**Here is the full syntax for tuning a decision tree**:

```python
 # Import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Instantiate a DecisionTreeClassifier 'dt'
dt = DecisionTreeClassifier(random_state=1)

# Print out 'dt's hyperparameters
print(dt.get_params())

 # Import GridSearchCV
from sklearn.model_selection import GridSearchCV

# Define the grid of hyperparameters 'params_dt'
param_grid = {
             'max_depth': [3, 4, 5, 6],
             'min_samples_leaf': [0.04, 0.06, 0.08],
             'max_features': [0.1, 0.3,0.5, 0.7]
}

# Instantiate a 10-fold CV grid search object 'grid_dt'
grid_dt= GridSearchCV(estimator=dt,
                       param_grid= param_grid,
                       scoring='accuracy',
                       cv=10,
                       n_jobs=8)

# Fit 'grid_dt' to the training data
grid_dt.fit(X_train, y_train)

# Extract best hyperparameters from 'grid_dt' and print them
best_hyperparams = grid_dt.best_params_

print('Best hyerparameters:\n', best_hyperparams)

# Extract best CV score from 'grid_dt'
best_CV_score = grid_dt.best_score_
print('Best CV accuracy'.format(best_CV_score))


# Extract best model from 'grid_dt'
best_model = grid_dt.best_estimator_

# Evaluate test set accuracy
test_acc = best_model.score(X_test,y_test)

# Print test set accuracy
print("Test set accuracy of best model: {:.3f}".format(test_acc))
```
