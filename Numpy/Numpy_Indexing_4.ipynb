# Numpy Subsetting (Indexing) [Slicing and dicing] 

```python
import numpy as np
```

### Indexing 1-D array

  - Indexing a one-dimensional array is equivalent to indexing a list
  
       - Use [] 
       
       - Use a single digit to get the element you want
       
       - Use [:] to select a range of points

```python
ar = np.array([0, 12, 21, 24, 5, 5, 22, 28, 10])
ar
```

```python
# The first element
ar[0]
```

```python
# The third element
ar[2]
```

```python
# The last element
ar[-1]
```

```python
# A range of elements: first up to third elements (third is included)
# remember the third element is indexed as 2, but we want to included so we put the index 3. 
ar[0:3]
```

#### Note: 

   - If we want from the beginning of an array up to a certain index, we use this syntax [:index]
   
   - If we want from a starting index to the last item we use this syntax [index:]

```python
# All first five elements, (0, 1, 2, 3, 4)

ar[:5]
```

```python
## Starting from the fifth element (fifth element is included so we use the index 4)

ar[4:]
```

```python
ar
```

### Note: 

   - We can use [:] to take all the elements from the array

```python
ar[:]
```

### Indexing using Step

   - Suppose we want to select elements using a step. To do that we use this syntax [start:stop:step]

```python
arr = np.array([*range(0, 11)])
arr
```

```python
# We can select elements from 0 to 9 with a step of two
arr[0:9:2]
```

```python
# Or we can use step of three
arr[:9:3]
```

### Reverse an Array

   - To reverse an array we can that using this syntax [::-1]

```python
arr[::-1]
```

### Note:

  - Indexing can be used for array re-assignment, we can change the content of an array by assing new value (s) to a subset

```python
ar2 = np.array([*range(7)])
ar2
```

```python
ar2[0:3] = 50
ar2
```

```python
ar2[-1] = 1000
ar2
```

### Warning:

  - Subbsetting in numpy does not make a copy of the object

```python
arr = np.arange(11)
arr
```

```python
first_half = arr[:6]
first_half
```

```python
## Let us reassing the sliced first half a new value
first_half[:] = 500
first_half
```

```python
## Let us print the original array
arr
```

   Not what we expected, huh!. That is how numpy does its work. It does not make a copy, but only refers to original object. Thus, when we change the subset, the original object changed as well, because they refer to the same object. That is memory efficient technique. And if you want a copy of an object you must do that explicitly.
   

### Copy method

```python
arr = np.arange(11)
arr_copy = arr.copy()
half_arr_copy = arr_copy[:6]
print(half_arr_copy)
half_arr_copy[:] = 100
print(half_arr_copy)
print(arr_copy)
print(arr)
```

### Indexing 2-D arrays

```python
from IPython.display import Image
Image("array_ind.png")
```

```python
arr_1 = [26, 23, 22, 27, 11]
arr_2 = [10, 11, 19, 21, 1]
arr_3 = [10, 0, 8, 18, 6]
mat = np.array([arr_1, arr_2, arr_3])
mat
```

#### What if we want to select a specific row

 suppose we want to select a specific row: Say the first one (**The First one is zero**) python in zero indexing, remember!!! 
   
   - we can do that by using [] square brackets

```python
mat.shape # (rows, columms)
```

```python
mat
```

```python
first_row = mat[0,]
first_row
```

```python
second_row = mat[1]
second_row
```

#### Note:

   - Selecting an entire row can be done in this way as well [row, : ] [row number , : ]

```python
row_one = mat[0, :]
row_one
```

```python
last_row = mat[-1, :]
last_row
```

#### What if we want to select a specific column
   
      - we can use this syntax [:, column], for example the first column would b [:,0]

```python
first_col = mat[:,0]
first_col
```

```python
second_col = mat[:, 1]
second_col
```

  - What if we need a specific value 
  
     - We can do that using double square brackets 
     
     - Or Using comma (,) like this [row, column]

```python
value_first_r_second_col = mat[0][1]
value_first_r_second_col
```

```python
# or to get the same
mat[0, 1]
```

### Selecting multiple rows

   - We use the colon (:)  [:, ] , [start_row **:** last_row (not included) **,** empty means all columns]

```python
first_and_second_row = mat[0:2, ]
first_and_second_row
```

### Selecting multiple columns

   - We use the colon (:) [,:] , [empty means all rows , start_col : last_col (not included)]

```python
first_and_second_col = mat[:, 0:2]
first_and_second_col
```

#### Selecting some rows and some columns

   - We want the first two rows, the second and the third columns
   
      - we can do that using this syntax [:, :] 

```python
sub = mat[0:2, 1:3]
sub
```

## Indexing Using Booleans
