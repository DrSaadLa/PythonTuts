### Creating Numpy Arrays from Lists

   - **A Numpy Array**:
   
       - It is an alternative to python list
       
       - It can perform calculations lists cannot do, such as **elementwise calculations**

```python
import numpy as np
# from numpy import array
```

```python
# Let us create an array from our previous lists 
np_height = np.array(height)
np_weight = np.array(weight)
```

```python
print(np_height)
print(np_weight)
```

```python
print(type(np_height))
print(type(np_weight))
```

### Solving the example

```python
# let us calculate bmi
%precision 3                   

bmi = np_weight / np_height**2
print(bmi)
```

```python
bmi
```

```python
print(type(bmi))
```
